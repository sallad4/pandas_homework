```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting


#Calculate the total number of schools
total_schools = school_data_complete['school_name'].nunique()
#Calculate the total number of students
total_kids = school_data_complete.drop_duplicates(['school_name','size']).agg({'size':'sum'})
#Calculate the total budget#
#total_budget = school_data_complete.drop_duplicates(['school_name','budget']).agg({'budget':'sum'}).map("${:,.2f}".format)
total_budget = school_data_complete['budget'].unique().sum()
#total_budget = '${:,.2f}'.format(total_budget)
#Calculate per student budget
per_student_budget = total_budget/total_kids
#Calculate the average math score
average_math_score = school_data_complete["math_score"].mean()
#Calculate the average reading score
average_reading_score = school_data_complete["reading_score"].mean()#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate = (average_math_score + average_reading_score) / 2
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = len(school_data_complete[school_data_complete['math_score']>= 70])/total_kids
#Calculate the percentage of students with a passing reading score (70 or greater)
big_readers = len(school_data_complete[school_data_complete['reading_score']>= 70])/total_kids
#Create a dataframe to hold the above results
frame_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": total_kids,
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Passing Rate": [passing_rate],
    "Passing Math": passing_math,
    "Passing Reading": big_readers
})
frame_df
```

```python
# Create an overview table that summarizes key metrics about each school, including:
# School Type
school_type = school_data_complete.groupby("school_name")[('type')].unique()
# Total Students
student_count = school_data_complete.groupby("school_name")[('size')].unique()
# Total School Budget
school_budget = school_data_complete.groupby("school_name")[('budget')].unique()
# Per Student Budget
per_student_dollars = (school_data_complete.groupby("school_name")[('budget')].unique() / student_count)
# Average Math Score
avg_math = school_data_complete.groupby('school_name')[("math_score")].mean()
# Average Reading Score
avg_reading = school_data_complete.groupby('school_name')[("reading_score")].mean()
# % Passing Math 
passing_math = (school_data_complete.loc[school_data_complete["math_score"] > 70, :].groupby('school_name')['math_score'].count() / student_count) * 100
# % Passing Reading
passing_reading = (school_data_complete.loc[school_data_complete["reading_score"] > 70, :].groupby('school_name')['reading_score'].count() / student_count) * 100
# Overall Passing Rate (Average of the above two)
overall = (passing_math + passing_reading) / 2

summary_df = pd.concat([school_type, 
           student_count, 
           school_budget, 
           avg_math, 
           avg_reading,
           per_student_dollars,
           passing_math, 
           passing_reading, 
           overall], 
          axis=1)
summary_df

summary_df.reset_index(inplace=True)

summary_df = summary_df.rename(columns={'index' : 'School Name',  
                                            'type' : 'School Type',
                                            'size' : 'Student Body',
                                            'budget' : 'School Budget',
                                            'math_score' : 'Average Math Score',
                                            'reading_score' : 'Average Reading Score',
                                            0 : 'Per Student Spending',                                                          
                                            1 : 'Passing Math Percent',
                                            2 : 'Passing Reading Percent', 
                                            3 : 'Overall Pass Rate'
                                        })
summary_df = summary_df.rename(columns={'school_name' : 'School Name'})

summary_df
```

```python
top_schools = summary_df.sort_values('Overall Pass Rate', ascending=False).iloc[0:5,]
top_schools
```

```python
#find the students by grade overall
frosh = school_data_complete.loc[school_data_complete['grade'] == '9th',]
soph = school_data_complete.loc[school_data_complete['grade'] == '10th',]
junior = school_data_complete.loc[school_data_complete['grade'] == '11th',]
senior = school_data_complete.loc[school_data_complete['grade'] == '12th',]
#count each grade by school
frosh_school_tot  = frosh.groupby('school_name')['grade'].count()
soph_school_tot = soph.groupby('school_name')['grade'].count()
junior_school_tot = junior.groupby('school_name')['grade'].count()
senior_school_tot = senior.groupby('school_name')['grade'].count()
#finding average math grade by year
frosh_math  = frosh.groupby('school_name')['math_score'].sum() / frosh_school_tot
soph_math = soph.groupby('school_name')['math_score'].sum() / soph_school_tot
junior_math = junior.groupby('school_name')['math_score'].sum() / junior_school_tot
senior_math = senior.groupby('school_name')['math_score'].sum() / senior_school_tot

math_by_grade = pd.concat([frosh_math, 
                                soph_math,
                                junior_math,
                                senior_math],
                                axis = 1)
math_by_grade

math_by_grade = math_by_grade.rename(columns={'school_name' : 'School Name',  
                                            0 : 'Freshmen',                                                          
                                            1 : 'Sophomores',
                                            2 : 'Juniors', 
                                            3 : 'Seniors'
                                        })

math_by_grade
```

```python
#finding average reading grade by year
frosh_reading  = frosh.groupby('school_name')['reading_score'].sum() / frosh_school_tot
soph_reading = soph.groupby('school_name')['reading_score'].sum() / soph_school_tot
junior_reading = junior.groupby('school_name')['reading_score'].sum() / junior_school_tot
senior_reading = senior.groupby('school_name')['reading_score'].sum() / senior_school_tot

reading_by_grade = pd.concat([frosh_reading, 
                                soph_reading,
                                junior_reading,
                                senior_reading],
                                axis = 1)
reading_by_grade

reading_by_grade = reading_by_grade.rename(columns={'index' : 'School Name',  
                                            0 : 'Freshmen',                                                          
                                            1 : 'Sophomores',
                                            2 : 'Juniors', 
                                            3 : 'Seniors'
                                        })
reading_by_grade
```

```python
spending_range = school_data_complete
spending_range['Per Student Spending'] = spending_range['budget'] / spending_range['size']

bins = [0, 585, 600, 625, 650, 675]
labels = ['<$585','$585-600','$600-625','$625-650','$650-675']
spending_range['Spending Per Student'] = pd.cut(spending_range["Per Student Spending"],bins, labels=labels)

per_budget = spending_range.groupby('Spending Per Student')['student_name'].count()

math_by_budget = spending_range.groupby('Spending Per Student')['math_score'].sum() / per_budget
read_by_budget = spending_range.groupby('Spending Per Student')['reading_score'].sum() / per_budget

passing_math = (school_data_complete.loc[school_data_complete["math_score"] > 70, :].groupby('Spending Per Student')['math_score'].count() / per_budget) * 100
passing_reading = (school_data_complete.loc[school_data_complete["reading_score"] > 70, :].groupby('Spending Per Student')['reading_score'].count() / per_budget) * 100
overall = (passing_math + passing_reading) / 2

does_spending_make_a_difference = pd.concat([math_by_budget,
                                 read_by_budget, 
                                 passing_math, 
                                 passing_reading,
                                 overall], 
                               axis = 1)
does_spending_make_a_difference = does_spending_make_a_difference.rename(columns={0 : 'Average Math Score',
                                                                                   1 : 'Average Reading Score',
                                                                                   2 : 'Passing Math Percent',
                                                                                   3 : 'Passing Reading Percent',
                                                                                   4 : 'Overall Pass Percent'      
                                                                                 })
does_spending_make_a_difference
```

```python
size_range = school_data_complete
size_range['Size Category'] = size_range['size']

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size_range['Size Category'] = pd.cut(size_range['Size Category'],size_bins, labels=group_names)

size_ish = size_range.groupby('Size Category')['student_name'].count()

math_by_size = size_range.groupby('Size Category')['math_score'].sum() / size_ish
read_by_size = size_range.groupby('Size Category')['reading_score'].sum() / size_ish

passing_math = (school_data_complete.loc[size_range["math_score"] > 70, :].groupby('Size Category')['math_score'].count() / size_ish) * 100
passing_reading = (school_data_complete.loc[size_range["reading_score"] > 70, :].groupby('Size Category')['reading_score'].count() / size_ish) * 100
overall = (passing_math + passing_reading) / 2

does_size_make_a_difference = pd.concat([math_by_size,
                                 read_by_size, 
                                 passing_math, 
                                 passing_reading,
                                 overall], 
                               axis = 1)

does_size_make_a_difference = does_size_make_a_difference.rename(columns={0 : 'Average Math Score',
                                                                                   1 : 'Average Reading Score',
                                                                                   2 : 'Passing Math Percent',
                                                                                   3 : 'Passing Reading Percent',
                                                                                   4 : 'Overall Pass Rate'
                                                                                 })
does_size_make_a_difference
```

```python
type_range = school_data_complete
type_range['School Type'] = type_range['type']

type_ish = type_range.groupby('School Type')['student_name'].count()

math_by_type = type_range.groupby('School Type')['math_score'].sum() / type_ish
read_by_type = type_range.groupby('School Type')['reading_score'].sum() / type_ish

passing_math = (school_data_complete.loc[type_range["math_score"] > 70, :].groupby('School Type')['math_score'].count() / type_ish) * 100
passing_reading = (school_data_complete.loc[type_range["reading_score"] > 70, :].groupby('School Type')['reading_score'].count() / type_ish) * 100
overall = (passing_math + passing_reading) / 2

does_type_make_a_difference = pd.concat([math_by_type,
                                 read_by_type, 
                                 passing_math, 
                                 passing_reading,
                                 overall],
                               axis = 1)
does_type_make_a_difference = does_type_make_a_difference.rename(columns={0 : 'Average Math Score',
                                                                                   1 : 'Average Reading Score',
                                                                                   2 : 'Passing Math Percent',
                                                                                   3 : 'Passing Reading Percent',
                                                                                   4 : 'Overall Pass Rate'
                                                                                 })
does_type_make_a_difference
```
